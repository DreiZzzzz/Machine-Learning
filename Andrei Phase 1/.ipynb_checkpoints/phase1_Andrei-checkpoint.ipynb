# How do fastest lap times vary across different race seasons, and are there any trends/decline in performance?

```python
# import the needed python packages and libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr # for 
from scipy.stats import spearmanr # for testing 
```

```python
# read and display csv
lap_times_df = pd.read_csv("lap_times.csv")
lap_times_df.head()
```

```python
# Since the lap_times_df doesn't have a column for the year/season of each race, we need to read another file to get the year/season of each observation.

# read and display csv
races_df = pd.read_csv("races.csv")
races_df.head()
```

```python
races_df.info()
```

```python
# Extract the raceId and year of the race from the races_df
# since other columns has a default value \N we cannot use those columns to create any further analysis, so we only extract raceID to be use as a matching index and the year of each of the race.
raceID_season_df = races_df[["raceId", "year"]]
raceID_season_df
```

```python
# check if the format of year is consistent
raceID_season_df["year"].unique()
```

```python
# check if the format of raceId is consistent
raceID_season_df["raceId"].unique()
```

```python
# feature engineering 

# add a column that converts milliseconds to seconds for easier analysis
lap_times_df["lap_time_in_seconds"] = lap_times_df["milliseconds"] / 1000
```

```python
# merge lap_times_df with raceID_season_df

lap_times_2_df = lap_times_df.merge(races_df[["raceId", "year"]], on="raceId", how="left")
lap_times_2_df.head()
```

```python
# Summary Statistics of lap_time per year
# group by year then get the summary statistics
grouped_df = lap_times_2_df.groupby("year")[["lap_time_in_seconds"]].agg(["mean", "median", "min", "max", "std"])
grouped_df
```

```python
# visualize using bar chart to easily compare statistics per year

grouped_df.plot(kind="bar", figsize=(14,6))
plt.xlabel("YEAR")
plt.ylabel("Lap Time (seconds)")
plt.title("Lap Time Statistics Per Year")
plt.legend(title="Key Statistics")
plt.xticks(rotation=45)
plt.show()
```

```python
# Check if there are missing values
lap_times_2_df.info()
```

```python
# Next is to find the fastest lap time per season

fastest_lap_per_year_df = lap_times_2_df.groupby("year")["lap_time_in_seconds"].min().reset_index()
fastest_lap_per_year_df
```

# Hypothesis 

## Null Hypothesis: There is no correlation between lap time and year
## Alternate Hypothesis : There is a correlation between lap time and year

```python
# test 1
# correlation = fastest_lap_per_year_df["lap_time_in_seconds"].corr(fastest_lap_per_year_df["year"])
# print("Correlation coefficient:", correlation)


# same as the one above but this method also provides the p_value
corr, p_value1 = pearsonr(fastest_lap_per_year_df["year"], fastest_lap_per_year_df["lap_time_in_seconds"])
print(f"Pearson Correlation coefficient: {corr}")
print(f"Pearson P-value: {p_value1}\n")

# try spearman correlation

correlation, p_value2 = spearmanr(fastest_lap_per_year_df["year"], fastest_lap_per_year_df["lap_time_in_seconds"])
print(f"Spearman correlation: {correlation}")
print(f"Spearman p-value: {p_value2}")




# Note:  if two different correlation tests, like Pearson and Spearman, yield similar or nearly 
# identical values (especially when accompanied by low p-values), it strongly suggests a 
# relationship exists between the two variables you're analyzing.
```

```python
# visualize using a scatter plot with a trend line.
sns.regplot(x="year", y="lap_time_in_seconds", data=fastest_lap_per_year_df, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Fastest Lap Time Per Year")
plt.show()
```

Conlusion: Since the calculated P-value is less than the 0.05 significance level, we reject the null hypothesis. This means that year and lap times are correlated.
